```python
#defining variables initially
'''
1. Define variable boat side as 'Right' initially
2. Create numeric variables for number of missionaries and cannibals
   on the righ and left side. So there are 3 missionaries and cannibals on right side and 0 on the left side 
'''
boat_side = "Right"
missionary_on_right = 3
cannibals_on_right = 3
missionary_on_left = 0
cannibals_on_left = 0

# Printing in a stylish manner like this
'''
M = 0 C = 0  |-----B|  M = 3 C = 3
'''
# Make sure to not hardcode all values.

print("M =", missionary_on_left, "C =", cannibals_on_left, end='')
print("  |-----B|  ", end='')
print("M =", missionary_on_right, "C =", cannibals_on_right)
```

```python
#Take input from user for deciding number of 
missionary = int(input("Enter number of Missionary in boat on "+ boat_side + ":"))
cannibals = int(input("Enter number of Cannibals in boat on " + boat_side + ":"))

#Maximum number of people one the boat should be 1 or 2
if (missionary+cannibals) != 1 and (missionary+cannibals) != 2:
    print("Invalid move")

#Turn based decisions
#first check boat side by checking its variable value
# use if else to check the side
if boat_side == "Right":
    #Inside the scope of if statement for right side
    
    ''' Now check if count for missionary or cannibal on the right
        side are less than the number that you entered for
        adding to boat if not print "Invalid move".'''
    if missionary > missionary_on_right or cannibals > cannibals_on_right:
        print("Invalid move")
        
    ''' Now when boad is on the right side after adding
    missionaries and cannibals in th boat they will 
    be reduced from the right side'''    
    missionary_on_right -= missionary
    cannibals_on_right -= cannibals
    
    ''' Similarly on the left side count for number of
        missionaries and cannibals will increase.'''
    missionary_on_left += missionary
    cannibals_on_left += cannibals
    
    ''' Display your game board in the same format as above like
    
        M = 0 C = 0  |-----B|  M = 3 C = 3

        Check the side of the Boat!!'''
    print("\nM =", missionary_on_left, "C =", cannibals_on_left, end='')
    print("  |B-----|  ", end='')
    print("M =", missionary_on_right, "C =", cannibals_on_right)
    
    '''Now we are done for the right side so lets change value of its variable'''
    boat_side = "Left"

elif boat_side == "Left":
    #Inside the scope of if statement for left side
    
    ''' Now check if count for missionary or cannibal on the right
    side are less than the number that you entered for
        adding to boat if not print "Invalid move".'''
    if missionary > missionary_on_left or cannibals > cannibals_on_left:
        print("Invalid move")
        
    ''' Now when boad is on the left side after adding
    missionaries and cannibals in th boat they will 
    be reduced from the left side'''    
    missionary_on_left -= missionary
    cannibals_on_left -= cannibals
    
    ''' Similarly on the right side count for number of
        missionaries and cannibals will increase.'''
    missionary_on_right += missionary
    cannibals_on_right += cannibals
    
    ''' Display your game board in the same format as above like
    
        M = 0 C = 0  |-----B|  M = 3 C = 3

        Check the side of the Boat!!'''
    print("\nM =", missionary_on_left, "C =", cannibals_on_left, end='')
    print("  |-----B|  ", end='')
    print("M =", missionary_on_right, "C =", cannibals_on_right)
    
    '''Now we are done for the right side so lets change value of its variable'''
    boat_side = "Right"
```

```python
'''
Now add conditions to check if the game is over:
1. If number of missionaries is less that number of cannibals on right side and
   number of missionaries are greater than zero.
2. If number of missionaries is less that number of cannibals on left side and 
   number of missionaries are greater than zero.
3. If over print 'YOU LOSE'
'''
#condition for losing
if (missionary_on_right != 0 and missionary_on_right < cannibals_on_right) or (missionary_on_left != 0 and missionary_on_left < cannibals_on_left):
    print("YOU LOSE")
'''
Add conditions to check if you have won the game:
1. Number of missionaries and cannibals on the left side should be equal to 3.
2. If condition matches print 'YOU WIN'.
'''    
#condition for winning
if missionary_on_left == 3 and cannibals_on_left == 3:
    print("YOU WIN")
```

```python
'''
The game must go on until you win or lose so lets put all conditions everything inside the infinite loop
1. Now for all conditions make appropriate changes
2. If move is invalid then without making any changes in the board you have to 'continue'.
   So make changes in all conditions
3. Now if you Win or Lose the game 'break' the loop and get out of it.
4. After getting out of the loop print "Game Over"
'''
## Copy display and variables part from lesson 1 for every iteration
from IPython.display import clear_output
boat_side = "Right"
missionary_on_right = 3
cannibals_on_right = 3
missionary_on_left = 0
cannibals_on_left = 0

print("M =", missionary_on_left, "C =", cannibals_on_left, end='')
print("  |-----B|  ", end='')
print("M =", missionary_on_right, "C =", cannibals_on_right)
## From here start the while infinite logic

while True:
    missionary = int(input("Enter number of Missionary in boat on "+ boat_side + ":"))
    cannibals = int(input("Enter number of Cannibals in boat on " + boat_side + ":"))
    
    #Maximum number of people one the boat should be 1 or 2
    if (missionary+cannibals) != 1 and (missionary+cannibals) != 2:
        print("Invalid move")
        continue
    
    #Turn based decisions
    if boat_side == "Right":
        if missionary > missionary_on_right or cannibals > cannibals_on_right:
            print("Invalid move")
            continue
           
        missionary_on_right -= missionary
        cannibals_on_right -= cannibals
        missionary_on_left += missionary
        cannibals_on_left += cannibals
        
        boat_side = "Left"
        clear_output()
        print("\nM =", missionary_on_left, "C =", cannibals_on_left, end='')
        print("  |B-----|  ", end='')
        print("M =", missionary_on_right, "C =", cannibals_on_right)
        
    else:
        if missionary > missionary_on_left or cannibals > cannibals_on_left:
            print("Invalid move")
            continue
            
        missionary_on_right += missionary
        cannibals_on_right += cannibals
        missionary_on_left -= missionary
        cannibals_on_left -= cannibals
        
        boat_side = "Right"
        clear_output()
        print("\nM =", missionary_on_left, "C =", cannibals_on_left, end='')
        print("  |-----B|  ", end='')
        print("M =", missionary_on_right, "C =", cannibals_on_right)
       
     
    #condition for losing
    if (missionary_on_right != 0 and missionary_on_right < cannibals_on_right) or (missionary_on_left != 0 and missionary_on_left < cannibals_on_left):
        print("YOU LOSE")
        break
         #condition for winning
    if missionary_on_left == 3 and cannibals_on_left == 3:
        print("YOU WIN")
        break
print("GAME OVER")
              
```
